```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
alpha_1 = 1
alpha_2 = 0.5
omega_0 = 0.002
omega_max = 1000
domega = 1e-3
tau = 20
```

```python
omega = np.arange(omega_0, omega_max, domega)
```

```python
def Sigma(s, tau, gamma):
    G0 = (omega_0**(1 - gamma)) / (gamma - 1)
    G1_of_s = np.sum((np.cos(omega*s) - 1) / omega**gamma) * domega
    G1_of_tau = np.sum((np.cos(omega*tau) - 1) / omega**gamma) * domega
    G2_of_s_tau = np.sum((np.cos(omega*s) - 1) * np.cos(omega*tau) / omega**gamma) * domega
    return 2*(alpha_1**2 + alpha_2**2) * (G1_of_s + G0) + 4*alpha_1*alpha_2*(G0 + G1_of_tau + G2_of_s_tau)
```

```python
s_ = np.arange(0, 60, 1)
Sigma_of_s_1 = [Sigma(s, tau, gamma=1.50) for s in s_]
Sigma_of_s_2 = [Sigma(s, tau, gamma=1.75) for s in s_]
Sigma_of_s_3 = [Sigma(s, tau, gamma=2.00) for s in s_]
```

```python
plt.plot(s_, (Sigma_of_s_1 - Sigma_of_s_1[0]))
plt.plot(s_, (Sigma_of_s_2 - Sigma_of_s_2[0]))
plt.plot(s_, (Sigma_of_s_3 - Sigma_of_s_3[0]))
```
